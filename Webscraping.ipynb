### Scraping the Data

Using the contents and `beautiful soup` load the data from the `By market capitalization` table into a `pandas` dataframe. The dataframe has the bank `Name` and `Market Cap (US$ Billion)` as column names.  Display the first five rows using head.


Using BeautifulSoup parse the contents of the webpage.


```python
#Replace the dots below
soup = BeautifulSoup(html_data,"html.parser")
soup_rows=soup.find_all('tr')
soup_rows
soup_row=soup_rows[0]
soup_row
print(type(soup_row))
soup_row.td
for i, row in enumerate(soup_rows):
    print("row",i,"is",row)
for i, row in enumerate(soup_rows):
    print("row",i)
    cells=soup_row.find_all('td')
    for j,cell in enumerate(cells):
        print('column',j,"cell",cell)
    
list_input=soup.find_all(name=["tr","td"])
list_input
soup.find_all(href="True")
soup.find_all(strings="Name")
soup.find("html_data", class_='Name')
table=soup.find('table')#in html table is represented by the tag <table>
len(table)
tables=soup.find_all('table')
len(tables)
for index,table in enumerate(tables): table_index = index
print(tables[table_index].prettify())


```

Load the data from the `By market capitalization` table into a pandas dataframe. The dataframe should have the bank `Name` and `Market Cap (US$ Billion)` as column names. Using the empty dataframe `data` and the given loop extract the necessary data from each row and append it to the empty dataframe.


```python

data=pd.DataFrame(columns=["Name","Market Cap(US$ Billion)"])

for row in tables[table_index].tbody.find_all("tr"):
    col=row.find_all("td")
    if (col!=[]):
        Name = col[1].text
        Marketcap = col[2].text 
        data=data.append({"Name":Name, "Market cap (US$Billion)": Marketcap}, ignore_index=True) # appending to th empty dataframe.
        data = data.replace(r'\n',' ', regex=True) #remove \n after each object in the cells of a dataframme
        data1 = data.dropna(how='all', axis='columns')#drop columns with Nan values
        
data1
```
